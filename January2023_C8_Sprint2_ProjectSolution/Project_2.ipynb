## Task-10:-

-- Sort thye data in descending order based on revenue and filter the to 500 movies.
-- Find the measures of central tendency of the following columns using the filtered data.
-- 1) Budget
-- 2) Revenue
-- 3) Runtime
-- Perform outlier analysis for the above three columns using box plots.

```python
sort_data= data.sort_values('revenue',ascending=False).head(500)
sort_data
```

```python
df=sort_data[['budget','revenue','runtime']]
df
```

```python
# Central Tendency
Mean=[]
Median=[]
Mode=[]

for col in df.head():
    Mean.append(df[col].mean())
    Median.append(df[col].median())
    Mode.append(df[col].mode()[0])
index=['Mean','Median','Mode']
cols=['budget','revenue','runtime']
```

```python
Central_tend=pd.DataFrame([Mean,Median,Mode],columns=cols,index=index)
Central_tend
```

```python
# Dispersion
Mean=[]
STD=[]
CV=[]

for col in df.head():
    Mean.append(df[col].mean())
    STD.append(df[col].std())
    CV.append(df[col].std()*100/df[col].mean())
    
index=['Mean','Standard Deviation','Coefficient of Variance']
cols=['budget','revenue','runtime']
```

```python
Measures_Desper=pd.DataFrame([Mean,STD,CV],columns=cols,index=index)
Measures_Desper
```

```python
# Outliers
for data in df.head():
    lo=[]                # lower outliers
    uo=[]                # Upper outliers
    ser=pd.Series(df[data])
    minimum=min(ser)
    maximum=max(ser)
    ranges=maximum-minimum
    q1=ser.quantile(0.25)
    q3=ser.quantile(0.75)
    iqr=q3-q1
    uf=q3+1.5*iqr
    lf=q1-1.5*iqr
    
    for i in df[data]:
        if i<lf:
            lo.append(i)
        elif i>uf:
            uo.append(i)
    if data=='budget':
        budget=['budget',minimum,maximum,ranges,q1,q3,iqr,lf,uf,(lo,uo)]
    elif data=='revenue':
        revenue=['revenue',minimum,maximum,ranges,q1,q3,iqr,lf,uf,(lo,uo)]
    elif data=='runtime':
        runtime=['runtime',minimum,maximum,ranges,q1,q3,iqr,lf,uf,(lo,uo)]
        
    
```

```python
outliers=pd.DataFrame([budget,revenue,runtime],columns=
                      ['data','minimum','maximum','ranges','q1','q3','iqr','lf','uf','Outliers'])
outliers
```

```python
# Converting to the list for boxplot

budget=df['budget'].tolist()
revenue=df['revenue'].tolist()
runtime=df['runtime'].tolist()
```

```python
fig,ax=plt.subplots(1,3,figsize=(15,5))
sns.boxplot(y=df['budget'],ax=ax[0], color="orange")
sns.boxplot(y=df['revenue'],ax=ax[2],color="teal")
sns.boxplot(y=df['runtime'],ax=ax[1],color="green")
plt.show()
```
